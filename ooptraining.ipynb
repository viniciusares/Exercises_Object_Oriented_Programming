```python
# Vinicius Emanoel Ares - Jun2019
# Classes and Objects training

class Restaurant():
    """this class stores restaurant's name and cuisine type"""
    def __init__(self, restaurant_name, cuisine_type, number_served=0):
        """Initialize the restaurant_name and cuisine_type attributes"""
        self.restaurant_name = restaurant_name
        self.cuisine_type = cuisine_type
        self.number_served = number_served
    def describe_restaurant(self):
        """Describe a restaurant by printing restaurant_name and cuisine_type"""
        print(self.restaurant_name.title() + " is a restaurant with a " + self.cuisine_type + " cuisine.")
    def open_restaurant(self):
        """Prints a message showing that the restaurant is open"""
        print(self.restaurant_name.title() + " is open.")
    def set_number_served(self):
        """docstring: Sets the number of customers served"""
        try:
            self.number_served = int(input("Enter the total number of customers served: "))
        except ValueError:
            print("Error: Invalid input. Only integers are accepted. Failed to perform action.")
    def increment_number_served(self):
        """increments the number served"""
        try:
            self.number_served += int(input("Increment the number of customers served by: "))
        except ValueError:
            print("Error: You must had entered an integer. Failed to perform action.")

class Foodtruck(Restaurant):
    """this subclass inherits from Restaurant class"""
    def __init__(self, restaurant_name, cuisine_type, number_served=0):
        super().__init__(restaurant_name, cuisine_type)
class IceCreamStand(Restaurant):
    def __init__(self, name, cuisine, flavors):
        super().__init__(name, cuisine)
        self.flavors = flavors
    def display_flavors(self):
        print(f'The {self.restaurant_name.title()} ice cream stand has the flavors: {self.flavors}')
        
class Users():
    def __init__(self, first_name, country, login_attempts=0):
        self.first_name = first_name
        self.country = country
        self.login_attempts = login_attempts
    def greet_user(self):
        print('welcome', self.first_name)
    def increment_login_attempts(self):
        """increment login attempts"""
        self.login_attempts += 1
    def reset_login_attemps(self):
        """sets the number of login attempts to zero"""
        self.login_attempts = 0
class Administrator(Users):
    def __init__(self, first_name, country, login_attempts=0):
        super().__init__(first_name, country)
class Privileges():
    def __init__(self, name):
        self.privileges = ['can add post', 'can delete post', 'can ban user',
                          'can promote user', 'can demote user']
        self.name = name
    def show_privileges(self):
        print(f'User {self.name} is an admin with privileges: {self.privileges}')
class Admin(Users):
    def __init__(self, name, country):
        super().__init__(name, country)
        self.privileges = Privileges(name)
        """self.privileges = ['can add post', 'can delete post', 'can ban user',
                          'can promote user', 'can demote user']
    def show_privileges(self):
        print(f'User {self.first_name} is an admin with privileges: {self.privileges}')"""

class Car():
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
class Battery():
    def __init__(self):
        self.capacity = 70
    def upgrade_battery(self):
        if self.capacity < 85:
            self.capacity = 85
        else:
            print('Battery is already the greater capacity')
class ElectricCar(Car):
    def __init__(self, make, model, year):
        super().__init__(make, model, year)
        self.battery = Battery()
    def get_range(self):
        if self.battery.capacity == 70:
            print('This car can go about 240 miles on a full charge')
        elif self.battery.capacity == 85:
            print('This car can go about 270 miles on a full charge')
```
